## 1.4.8 규모구분을 전용면적 컬럼으로 변경
규모구분 컬럼은 전용면적에 대한 내용이 있습니다. 전용면적이라는 문구가 공통적으로 들어가고 규모구분보다는 전용면적이 좀 더 직관적이기 때문에 전용면적이라는 컬럼을 새로 만들어주고 기존 규모구분의 값에서 전용면적, 초과, 이하 등의 문구를 빼고 간결하게 만들어 봅니다.

이때 str의 replace 기능을 사용해서 EX) "전용면적 60 초과 85 이하"라면 "60m~85m"로 변경해줍니다.

```python
df_last
```

```python
df_last["규모구분"].unique()
```

```python
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"]
```

## 1.8.1 melt로 Tidy data 만들기

pandas의 melt를 사용하면 데이터 형태를 변경할 수 있다. df_first 변수에 담긴 데이터프레임은 df_last에 담겨있는 데이터프레임의 모습과 다릅니다. 같은 형태로 만들어주어야 데이터를 합칠 수 있다. 데이터를 병합하기 위해 melt를 사용해 열에 있는 데이터를 행으로 녹여봅니다.

- http://pandas.pydata.org/docs/user_guide/reshaping.html#reshaping-by-melt
- Tidy Data란?

```python
df_first.head(1)
```

```python
# 데이터 녹여서 새로 만들기

df_first_melt = df_first.melt(id_vars="지역", var_name="기간", value_name="평당분양가격")
df_first_melt.head()
```

```python
df_first_melt.columns = ["지역명", "기간", "평당분양가격"]
df_first_melt.head()
```

## 1.8.2 연도와 월을 분리하기 

- pandas string-handling 사용하기: 

```python
date = "2013년12월"
date
```

```python
# split을 통해 "년"을 기준으로 분리
date.split("년")[0]
```

```python
date.split("년")[0]
```

```python
date.split("년")[-1].replace("월", "")
```

```python
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year

y = parse_year(date)
print(type(y))
y
```

```python
parse_year(date)
```

```python
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)
    return month

# 함수 정의 단계

parse_month(date)
```

```python
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt.head
```

```python
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```
